```python
# Group By operation

my_data = {'Name': ['Jai','Anuj','Jai','Prince','Gaurav','Anuj','Prince','Abhi'],
           'Age': [27,24,22,32,33,36,27,32],
           'Address':['Nagpur','Kanpur','West Bengal','Kanpur','Allahbad','Nagpur','Aligarh','Delhi'],
           'Qualification': ['Bsc','Mtech','BCA','MCA','Btech','Mtech','Phd','Btech']
}
df_new = pd.DataFrame(my_data)
df_new
```

```python
# Group by Name

gk = df_new.groupby('Name').groups
gk
```

```python
gk = list(df_new.groupby('Name'))
gk
```

```python
# Group by using multiple keys
df_new
```

```python
df_new = df_new.groupby(['Name','Qualification'])
print(df_new.groups)
```

```python
list(df_new)
```

```python
# Merge two dataframes
my_data1 = {'Name': ['Jai','Anuj','Jai','Prince'],
           'Age': [27,24,22,32],
           'Address':['Nagpur','Kanpur','West Bengal','Kanpur'],
           'Qualification': ['Bsc','Mtech','BCA','MCA']
} 

my_data2 = {'Name': ['Gaurav','Anuj','Prince','Abhi'],
           'Age': [33,36,27,32],
           'Address':['Allahbad','Nagpur','Aligarh','Delhi'],
           'Qualification': ['Btech','Mtech','Phd','Btech']
}
```

```python
df_data1 = pd.DataFrame(my_data1)
df_data2 = pd.DataFrame(my_data2)
```

```python
df_data1
```

```python
df_data2
```

```python
df_data3 = pd.concat([df_data1,df_data2])
df_data3
```
